```python


# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'x': np.random.normal(0, 1, 500),
    'y': np.random.normal(0, 1, 500),
    'category': np.random.choice(['A', 'B', 'C', 'D'], size=500)
})

# Create a scatter plot with color encoding and tooltips
scatter = alt.Chart(data).mark_circle().encode(
    x='x',
    y='y',
    color='category',
    tooltip=['x', 'y', 'category']
).properties(
    width=500,
    height=400
)

# Create a histogram of x values
histogram = alt.Chart(data).mark_bar().encode(
    x=alt.X('x', bin=alt.BinParams(maxbins=30)),
    y='count()',
    color=alt.value('steelblue')
).properties(
    width=500,
    height=150
)

# Create a vertical line for average x value
average_line = alt.Chart(data).mark_rule().encode(
    x='mean(x)',
    color=alt.value('red')
)

# Combine the scatter plot, histogram, and average line
chart = (scatter | (histogram + average_line)).resolve_scale(y='independent')

chart
```

```python

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'x': np.random.normal(0, 1, 100),
    'y': np.random.normal(0, 1, 100),
    'category': np.random.choice(['A', 'B'], size=100)
})

# Scatter plot with color encoding
scatter = alt.Chart(data).mark_circle().encode(
    x='x',
    y='y',
    color='category',
    tooltip=['x', 'y', 'category']
).properties(
    width=300,
    height=300
)

# Histogram of x values
histogram_x = alt.Chart(data).mark_bar().encode(
    x=alt.X('x', bin=alt.BinParams(maxbins=30)),
    y='count()',
    color=alt.value('steelblue')
).properties(
    width=300,
    height=150
)

# Histogram of y values
histogram_y = alt.Chart(data).mark_bar().encode(
    x='count()',
    y=alt.Y('y', bin=alt.BinParams(maxbins=30)),
    color=alt.value('steelblue')
).properties(
    width=150,
    height=300
)

# Kernel density plot of x and y values
density = alt.Chart(data).transform_density(
    density='x',
    as_=['x', 'density']
).mark_line().encode(
    x='x:Q',
    y='density:Q',
    color=alt.value('red')
).properties(
    width=300,
    height=150
)

# Combine the subplots
chart = alt.vconcat(
    scatter | histogram_x,
    histogram_y & density
).resolve_scale(y='shared')

chart
```

```python
import altair as alt
import pandas as pd
import numpy as np

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'x': np.random.normal(0, 1, 100),
    'y': np.random.normal(0, 1, 100),
    'category': np.random.choice(['A', 'B', 'C'], size=100)
})

# Define the number of rows and columns for subplots
num_rows = 2
num_cols = 3

# Initialize an empty list to store the individual charts
charts = []

# Loop through the rows and columns
for row in range(num_rows):
    row_charts = []
    for col in range(num_cols):
        # Create a scatter plot for each subplot
        scatter = alt.Chart(data).mark_circle().encode(
            x='x',
            y='y',
            color='category',
            tooltip=['x', 'y', 'category']
        ).properties(
            width=200,
            height=150
        )
        # Append the scatter plot to the row_charts list
        row_charts.append(scatter)
    
    # Concatenate the charts horizontally for each row
    row_chart = alt.hconcat(*row_charts)
    # Append the row_chart to the charts list
    charts.append(row_chart)

# Concatenate the row charts vertically to create the final subplot layout
subplot = alt.vconcat(*charts)

subplot
```

```python
import altair as alt
from vega_datasets import data

# Load a sample dataset
source = data.cars()

# Define the number of rows and columns for subplots
num_rows = 2
num_cols = 3

# Calculate the total number of subplots
total_subplots = num_rows * num_cols

# Get unique values from the "Origin" column
unique_origins = source['Origin'].unique()

# Select the first n unique origins to match the number of subplots
unique_origins = unique_origins[:total_subplots]

# Initialize an empty list to store the charts
charts = []

# Loop through the subplots
for i, origin in enumerate(unique_origins):
    # Calculate the row and column index for the current subplot
    row = i // num_cols
    col = i % num_cols

    # Create a chart for the current subplot
    chart = alt.Chart(source).mark_point().encode(
        x=alt.X('Horsepower:Q', title='Horsepower'),
        y=alt.Y('Miles_per_Gallon:Q', title='Miles per Gallon'),
        color='Origin:N'
    ).transform_filter(
        alt.datum.Origin == origin  # Filter data based on subplot origin
    ).properties(
        width=200,
        height=150,
        title=origin  # Set the subplot title
    )

    # Add the chart to the list
    charts.append(chart)

# Concatenate the charts into a grid layout
grid_chart = alt.vconcat(*[alt.hconcat(*charts[i:i + num_cols]) for i in range(0, total_subplots, num_cols)])

grid_chart
```
